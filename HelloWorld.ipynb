```python
#Step 1 - importing data
import pandas as pd
 #the below sklearn line from the below code can be written here to make the code look organized
music_data = pd.read_csv('music.csv')   
music_data

#step 2 - Preparing/Cleaning the data   #like removing dublicates , none values and so on
X = music_data.drop(columns=['genre'])   # input set#this is to separate the data so that we can separate the data into input and output - by convention we use X
X   #now you have data without genre , which is our output or predections that we are trying to do
y = music_data['genre']   #output set  - we use lowercase y to do this - by convention
y

#step 3 - Creating model using ML algorithm (Learning and Predicting)
    #we will be using simple algorithm called decision tree from a librabry called scikit learn
from sklearn.tree import DecisionTreeClassifier  #sklearn is a package from scikit learn library and tree is the modul and we have class claaed DesictionTreeCLassifier

# our_model= DecisionTreeClassifier()        #creating new instance of the class
# our_model.fit(X, y) #this takes 2 dataset input set its X and output set - its y
# predections = our_model.predict([  [21, 1] , [22, 0]   ])  #it takes two dimensional array means Array inside an Array #here we are asking to make two predictions 1. 21 year old man likes what? and 22 year old female likes what??
# predections

#Step 4 Calculating Accuracy of our model- Always measure Accuracy of the Model that we have built
    #Allocate some data for training and some for Testing
    #Rule of Thumb is ... 70% to 80% data should be allocated for Training and Rest should be for Testing
    #So Above we just used two dataset , which is not going to measure Accuracy of our model
from sklearn.model_selection import train_test_split  #with this function we can easily split our dataset into two sets  Training and Testing
X_train, X_test , y_train, y_test = train_test_split(X, y , test_size = 0.2) # We are allocating 20% of our data to Testing - This function returns a Tuple - so we can unpack it to full variable hear only

model = DecisionTreeClassifier()
model.fit(X_train,y_train)   # X_train for input and y_train for outputdata
predections = model.predict(X_test) #X_test taken as input dataset 

from sklearn.metrics import accuracy_score

score = accuracy_score(y_test, predections)

score        # 1 means 100% and every time you may see a diffent score

from sklearn.externals import joblib        #this function is used when we dont want to Train the model every time there is new information 
                        #joblib has methods for saving and loading models
                        
                        
joblib.dump(model, 'music-recommender.joblib)  #two arguments our model and name of the file we want to store into 

```

```python
#TO visualize Data

import pandas as pd
from sklearn.tree import DecisionTreeClassifier  #sklearn is a package from scikit learn library and tree is the modul and we have class claaed DesictionTreeCLassifier
from sklearn import tree #this has method to extract our decision tree to export in graphical formate

music_data = pd.read_csv('music.csv')

X = music_data.drop(columns=['genre'])   # input set#this is to separate the data so that we can separate the data into input and output - by convention we use X
y = music_data['genre']   #output set  - we use lowercase y to do this - by convention

model = DecisionTreeClassifier()  #Create Model
model.fit(X, y) #Train the Model

tree.export_graphviz(model, out_file = 'music-recommender.dot',
                     feature_names= ['age', 'gender'],
                     class_names = sorted(y.unique()),
                     label = 'all',
                     rounded = True,
                     filled = True)   #arguments we need to pass - model and name of the output file(this we are passing as keyword and argument) unique as its repeated
                    

 
```
